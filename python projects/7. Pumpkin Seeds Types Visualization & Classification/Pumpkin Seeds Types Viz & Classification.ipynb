<h1 style="background-color:#f0d8b6;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">Pumpkin Classification </h1>

Pumpkins, widely grown for food and ornamental purposes, come in various species and varieties, each with unique seed characteristics. These seeds can be categorized based on their physical properties, such as size, shape, and texture. The ability to classify pumpkin seeds can have practical applications in agriculture, seed quality assessment, and even in the food industry, where certain seed types may be preferred for specific products.

### <b><span style='color:#F5B04B'>|</span> Dataset Overview</b>

The Pumpkin Seeds Dataset is a comprehensive collection of data that includes measurements and characteristics of different pumpkin seed varieties. The dataset is designed to facilitate classification tasks in the field of agriculture, particularly in distinguishing between different types of pumpkin seeds. 

Based on common definitions used in image processing and geometric analysis of objects, especially in the context of biological shapes like seeds. These definitions are derived from general knowledge of how these geometric properties are usually calculated and interpreted in scientific and machine learning contexts. Here's a brief explanation of each:

### <b><span style='color:#F5B04B'>|</span> Features</b>

<br>

1. **`Area` [numeric]:** The total area of the pumpkin seed in pixels.
2. **`Perimeter` [numeric]:** The perimeter length of the seed boundary.
3. **`Major_Axis_Length` [numeric]:** The length of the major axis of the seed, representing the longest diameter.
4. **`Minor_Axis_Length` [numeric]:** The length of the minor axis of the seed, representing the shortest diameter.
5. **`Convex_Area` [numeric]:** The area of the convex hull around the seed, which is the smallest convex shape that can enclose the seed.
6. **`Equiv_Diameter` [numeric]:** The diameter of a circle with the same area as the seed.
7. **`Eccentricity` [numeric]:** A measure of how elongated the seed is, calculated based on the major and minor axis lengths.
8. **`Solidity` [numeric]:** The proportion of the convex area that is filled by the seed (Area/Convex_Area).
9. **`Extent` [numeric]:** The ratio of the seed's area to the area of the bounding box (a rectangle enclosing the seed).
10. **`Roundness` [numeric]:** A measure of how close the shape of the seed is to a perfect circle.
11. **`Aspect_Ratio` [numeric]:** The ratio of the major axis length to the minor axis length.
12. **`Compactness` [numeric]:** A measure combining area and perimeter to reflect how compact the seed is.

<br>

### <b><span style='color:#F5B04B'>|</span> Target Variable</b>

<br>

**`Class` [categorical]:** The type or class of the pumpkin seed, which is the target variable for classification.

### <b><span style='color:#F5B04B'>|</span> Univariate Analysis</b>

```python
def plot_distribution(df, feature, plot_title):
    with plt.style.context('seaborn-whitegrid'):
        fig, ax = plt.subplots(1, 2, figsize=(20, 5))

        # Colors for pumpkin seeds theme
        hist_color = '#F5B04B'  # Pumpkin Orange
        box_color = '#F5B04B'   # Autumn Brown

        # Histogram
        sns.histplot(x=feature,
                     data=df, 
                     ax=ax[0],
                     linewidth=0.7,
                     color=hist_color,
                     bins=20).set_title(f"Distribution of {plot_title}")
        ax[0].set_xlabel(plot_title)

        # Box plot for the entire distribution
        sns.boxplot(x=feature,
                    data=df,
                    linewidth=1.4,
                    width=0.5,
                    color=box_color,
                    ax=ax[1]).set_title(f"Box Plot of {plot_title}")
        ax[1].set_xlabel(plot_title)

        plt.show()
```

### <b><span style='color:#F5B04B'>|</span> Train Test Split</b>

```python
X = df[['Area', 'Perimeter', 'Major_Axis_Length', 'Minor_Axis_Length',
       'Convex_Area', 'Equiv_Diameter', 'Eccentricity', 'Solidity', 'Extent',
       'Roundness', 'Aspect_Ration', 'Compactness']]      # Features

Y = df['Class']                    # Target

# Train_Test Split
x_train , x_test , y_train , y_test = train_test_split(X, Y, test_size=0.2, random_state=21, shuffle = True, stratify=Y)
```

### <b><span style='color:#F5B04B'>|</span> Building the modeling & evaluation function</b>

```python
def evaluate_classifier(name, classifier, params, X_train=x_train, y_train=y_train, X_test=x_test, y_test=y_test):
    # Perform grid search with cross-validation
    grid_search = GridSearchCV(estimator=classifier, param_grid=params, cv=10)
    grid_search.fit(X_train, y_train)

    # Best estimator from grid search
    best_model = grid_search.best_estimator_

    # Predict the target variable for the test set
    y_pred = best_model.predict(X_test)

    # Compute and print accuracy
    accuracy = best_model.score(X_test, y_test)
    print(f'Classifier: {name}')
    print(f'Best Parameters: {grid_search.best_params_}')
    print(f'Accuracy: {round(accuracy, 2)}')

    # Compute additional performance metrics
    cm = confusion_matrix(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    
    # Class labels dictionary
    class_labels = {'erevelik': 'erevelik', 'rgp Sivrisi': 'rgp Sivrisi'}

    print(f'F1 Score: {round(f1, 2)}')
    print(f'Precision: {round(precision, 2)}')
    print(f'Recall: {round(recall, 2)}')
    print('Classification Report:')
    print(classification_report(y_test, y_pred, target_names=[class_labels[label] for label in np.unique(y_test)]))
    
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    plt.imshow(cm, interpolation='nearest', cmap='Oranges')
    plt.title('Confusion Matrix')
    plt.colorbar()
    plt.xticks(np.arange(len(class_labels)), [class_labels[label] for label in class_labels], rotation=45)
    plt.yticks(np.arange(len(class_labels)), [class_labels[label] for label in class_labels])
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')

    # Add numbers to the confusion matrix
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, str(cm[i, j]), horizontalalignment='center', verticalalignment='center', color='#F5B04B')
    
    plt.tight_layout()
    plt.show()
```

### <b><span style='color:#F5B04B'>|</span> Logistic Regression</b>

```python
# Define the logistic regression classifier with its corresponding hyperparameters
logistic_regression_params = {
    'C': [0.01, 0.1, 1, 10],                 
    'penalty': ['l2'],                        
    'max_iter': [100, 200, 500],             
    'class_weight': [None, 'balanced']
}

logistic_regression_classifier = LogisticRegression()

# Use the evaluate_classifier function to evaluate the logistic regression classifier
evaluate_classifier('Logistic Regression', logistic_regression_classifier, logistic_regression_params)
```

### <b><span style='color:#F5B04B'>|</span> KNN</b>
